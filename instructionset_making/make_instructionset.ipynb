```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
import json
```

```python
data = pd.read_csv("/data8/auto-prompt/self_introduction_evaluator/instructionset_making/recodings_samsung_2016.csv")
```

```python
evaluation_dict = {'삼성전자의 비전, 가치관, 사업 영역에 대한 이해 및 공감':0, '지원자의 장기적인 목표와 성장의지를 바탕으로 한 성장가능성':0,'직무 수행에 대한 구체적인 계획':0, '지원자의 성과에 대한 자신감 및 열의':0, '삼성전자의 인재상과 지원자의 가치관, 기업의 성장 배경과 지원자의 성장 배경 간 일치도':1, '문제해결력 및 도전정신, 회복 탄력성':1, '윤리적 사고 및 성실함과 책임감':1, '협업 및 의사소통능력':1, '사회 이슈에 대한 이해도':2, '삼성전자의 사업 방향에 대한 이해 및 기여 가능성':2, '논리적 사고력':2, '다양성에 대한 이해와 포용력':2}
```

```python
question_dict = {0:"삼성전자를 지원한 이유와 입사 후 회사에서 이루고 싶은 꿈을 기술하십시오. (700자 이내)", 
                 1:"본인의 성장과정을 간략히 기술하되 현재의 자신에게 가장 큰 영향을 끼친 사건, 인물 등을 포함하여 기술하시기 바랍니다. (※작품속 가상인물도 가능) (1500자)", 
                 2:"최근 사회이슈 중 중요하다고 생각되는 한가지를 선택하고 이에 관한 자신의 견해를 기술해 주시기 바랍니다. (1000자)"}
```

```python
#자소서와 평가지표별 승자 개수 세기
### 실제로 할 떄는 46으로
data_list = []

for i in range(46):
    for k in evaluation_dict.keys():
        count_num = data[(data['평가지표'] == k) & (data['승자'] == str(i))]['승자'].count()
        data_list.append([k, i, count_num])

df = pd.DataFrame(data_list, columns=['평가지표', '자소서', 'count_num'])
```

```python
df
```

```python
# dict_info 가져오기
json_file_path = '/data8/auto-prompt/self_introduction_evaluator/instructionset_making/data_info.json'

with open(json_file_path, 'r', encoding='utf-8') as file:
    dict_info = json.load(file)

for key, value in dict_info.items():
    print(f"{key}: {value}")
```

```python
#평가지표를 문항별로 합치기

df['문항'] = df['평가지표'].map(evaluation_dict)
result = df.groupby(['자소서', '문항'])['count_num'].sum()

result
```

```python
#딕셔너리로 저장
result_dict = result.to_dict()
print(result_dict)
```

```python
final_dict = {}

# 항목별 자소서마다 에세이는 dict_info에서, question은 question_info에서, 평가항목들은 evaluation_metric과 df 데이터프레임에서 가져와서 딕셔너리로 저장
for key, value in result_dict.items():
    essay_num, eval_group = key
    new_key = f"{essay_num}_{eval_group}"
    essay_info = dict_info[new_key]
    question_info = question_dict[eval_group]
    eval_item = [(k, df[(df['평가지표'] == k) & (df['자소서'] == essay_num)]['count_num'].values[0]) for k, v in evaluation_dict.items() if v == eval_group]
    final_dict[new_key] = {'essay_info': essay_info, 'question_info':question_info, 'eval_item': eval_item, 'count_num': value}

print(final_dict)
```

```python
#instruction, input, ouput형태의 jsonl로 변환

with open('instruction_tuning_train_data.jsonl', 'w', encoding='utf-8') as f:
    for key, value in final_dict.items():
        essay_info = value['essay_info']
        question_info = value['question_info']
        eval_item = value['eval_item']
        count_num = value['count_num']
        output= f"총 점수는 {count_num}점 입니다."
        for evaluation in eval_item:
            output += f"{evaluation[0]} 평가지표는 {evaluation[1]}점, "
        output = output[:-2]
        output += "이기 때문입니다."
        dict_to_write = {
            "instruction": f"{question_info}에 대하여 작성한 에세이의 점수를 매겨주세요",
            "input": essay_info,
            "output": output,
            "score": count_num}
        json.dump(dict_to_write, f, ensure_ascii=False)
        f.write('\n')
```
