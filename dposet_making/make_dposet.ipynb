```python
import pandas as pd
# read 'instruction_tuning_train_data.jsonl' and make dataframe
df = pd.read_json('instruction_tuning_train_data.jsonl', lines=True)
```

```python
df['score']
```

```python
sorted_df = df.sort_values(by='score', ascending=False)
```

```python
A = sorted_df[0:50]
B = sorted_df[-50:]
```

```python
A
```

```python
len(B)
```

```python
data_list = []
for num_A in range(len(A)):
    for num_B in range(len(B)):
        data_list.append([A.iloc[num_A]['instruction'], A.iloc[num_A]['input'], A.iloc[num_A]['output'], B.iloc[num_B]['instruction'], B.iloc[num_B]['input'], B.iloc[num_B]['output']])
        
C = pd.DataFrame(data_list, columns=['instruction_A', 'input_A', 'output_A', 'instruction_B', 'input_B', 'output_B'])
```

```python
C.to_csv('DPOset.csv', index=False)
```
