```python
import pandas as pd
# 문항, 평가항목의 dataframe 생성
df = pd.DataFrame(columns=['question', 'grading'])
```

```python
question1 = '삼성전자를 지원한 이유와 입사 후 회사에서 이루고 싶은 꿈을 기술하십시오. (700자 이내)'
question2 = '본인의 성장과정을 간략히 기술하되 현재의 자신에게 가장 큰 영향을 끼친 사건, 인물 등을 포함하여 기술하시기 바랍니다. (※작품속 가상인물도 가능) (1500자)'
question3 = '최근 사회이슈 중 중요하다고 생각되는 한가지를 선택하고 이에 관한 자신의 견해를 기술해 주시기 바랍니다. (1000자)'
```

```python
grading1 = ['삼성전자의 비전, 가치관, 사업 영역에 대한 이해 및 공감',
            '지원자의 장기적인 목표와 성장의지를 바탕으로 한 성장가능성',
            '직무 수행에 대한 구체적인 계획',
            '지원자의 성과에 대한 자신감 및 열의']

grading2 = ['삼성전자의 인재상과 지원자의 가치관, 기업의 성장 배경과 지원자의 성장 배경 간 일치도',
            '문제해결력 및 도전정신, 회복 탄력성',
            '윤리적 사고 및 성실함과 책임감',
            '협업 및 의사소통능력']

grading3 = ['사회 이슈에 대한 이해도',
            '삼성전자의 사업 방향에 대한 이해 및 기여 가능성',
            '논리적 사고력',
            '다양성에 대한 이해와 포용력']
 
```

```python
# df에 문항, 평가항목 추가
df.loc[0] = [question1, grading1]

df.loc[1] = [question2, grading2]

df.loc[2] = [question3, grading3]
 
```

```python
# df를 딕셔너리로 저장
df_dict = df.to_dict()
df_dict

# 문항, 평가항목을 json 파일로 저장
import json
with open('/data8/auto-prompt/self_introduction_evaluator/guides/grading_2016.json', 'w') as f:
    json.dump(df_dict, f, indent=4)
```
